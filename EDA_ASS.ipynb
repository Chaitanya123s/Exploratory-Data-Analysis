```python
# prompt: ●Use bar charts  to display the frequency of categories for categorical variables.

# Categorical variables to plot
categorical_cols = ['NSP', 'Tendency'] # Assuming 'NSP' and 'Tendency' are categorical

for col in categorical_cols:
    plt.figure(figsize=(8, 6))
    sns.countplot(x=col, data=pd, palette='viridis')
    plt.title(f'Frequency of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

```python
# Generate scatter plots for a few pairs of variables to explore relationships
sns.scatterplot(x='LB', y='AC', data=pd)
plt.title('Scatter plot of LB vs AC')
plt.show()

sns.scatterplot(x='LB', y='FM', data=pd)
plt.title('Scatter plot of LB vs FM')
plt.show()

# Generate a correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pd.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
